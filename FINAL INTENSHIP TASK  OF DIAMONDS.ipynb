```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("diamonds.csv")
df
```

# 1.  Print the total number of entries in the dataset


```python
print("Total number of entries:", len(df))
```

# 2. Plot the distribution of diamond prices.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, color='skyblue')
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#  3. Calculate the average price for each cut and color.


```python
average_prices = df.groupby(['cut', 'color'])['price'].mean().reset_index()

print(average_prices)
```

# 4.Print the names of diamonds with the highest and lowest prices

```python
highest_price_row = df.loc[df['price'].idxmax()]

```

```python
lowest_price_row = df.loc[df['price'].idxmin()]
```

```python
print("Diamond with the highest price:", highest_price_row['price'])
print("Diamond with the lowest price:", lowest_price_row['price'])
```

#  5.Plot the average price by cut and color.


```python
average_prices = df.groupby(['cut', 'color'])['price'].mean().reset_index()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=average_prices, x='cut', y='price', hue='color')
plt.title('Average Price by Cut and Color')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.show()
```

# 6.Calculate the correlation between diamond carat weight and price.


```python
correlation = df['carat'].corr(df['price'])

print("Correlation between carat and price:", correlation)
```

# 15. Plot the average price by cut and clarity.


```python
average_price = df.groupby(['cut', 'clarity'])['price'].mean().reset_index()
```

```python
pivot_table = average_price.pivot(index='cut', columns='clarity', values='price')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".2f")
plt.title('Average Price by Cut and Clarity')
plt.xlabel('Clarity')
plt.ylabel('Cut')
plt.show()
```
